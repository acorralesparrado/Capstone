```python
import pandas as pd
import numpy as np
```

```python
#Loading the table
df=pd.read_html("https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050.")[0]
```

```python
df.head()
```

```python
#Drop "Not Assigned" from "Borough"
df.drop(df.loc[df['Borough']=='Not assigned'].index, inplace=True)
df.head(5)
```

```python
#Merge "Neighbourhood" in the same "Borough"
df_tor = df.astype(str).groupby('Postcode').agg(lambda x: ','.join(x.unique()))
df_tor.reset_index(inplace = True) 
df_tor.head(5)
```

```python
#Clean "Neighbourhood" from "Not Assigned"
df_tor['Neighbourhood'].replace("Not assigned", np.nan, inplace = True)
df_tor.Neighbourhood.fillna(df_tor.Borough, inplace=True)
df_tor.head(10)
```

```python
#Save to CSV
df_tor.to_csv('Toronto_PostalCodes.csv')
```

```python
#Sort by "Neighbourhood"
toronto_set = df_tor.sort_values(by=['Neighbourhood'])
toronto_set.head(5)
```

```python
#Shape
toronto_set.shape
```

```python
# Load Geospatial Data url = 'https://cocl.us/Geospatial_data'
df_cord = pd.read_csv('Geospatial_Coordinates.csv')
df_cord.head(5)
```

```python
#Merge both DF
df_toronto = pd.concat([df_tor, df_cord], axis = 1)
```

```python
#Final DF
df_toronto.head()
```

```python
#Tidy - Drop Postal Code column
df_toronto.drop('Postal Code', axis=1, inplace = True)
df_toronto.head(10)
```

```python
import folium
from geopy.geocoders import Nominatim
```

```python
#Get Toronto Coordinates
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# Extracting Borough
df_toronto.head()
Torontogeodf = df_toronto.loc[df_toronto['Borough'].str.contains('Toronto')].reset_index()
Torontogeodf.shape
Torontogeodf.head()
```

```python
# Map of Toronto 
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# Adding Markers to map
for lat, lng, borough, neighborhood in zip(Torontogeodf['Latitude'], Torontogeodf['Longitude'], Torontogeodf['Borough'], Torontogeodf['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
